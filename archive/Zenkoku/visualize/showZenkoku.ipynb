# 上位N件リポジトリの名前とキーワード取得

```python
# 上位Nリポジトリのキーワード収集
def getKeywordsTOP5(codedf, N):
  print('fast-slow上位' + str(N) + 'リポジトリ')
  print(codedf.head(N))
  # print()

  i = 0
  keywords = []
  repo = []
  for row in codedf.itertuples():
    i += 1
    # print(str(i) + ' ' + row.repository)
    repo.append(row.repository)
    key = df_keywords[df_keywords['repository'] == row.repository]['keywords']
    for k in key:
      # print(len(k))
      keywords.extend(k)
    # print()

    # 上位n件
    if (i == N):
      # print(len(keywords))
      keywords_count = Counter(keywords).most_common()
      break

  return keywords_count, repo
```

```python
# 色わけ

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def create_log_scatter_plot(df_date, repo_date=[]):
    """
    対数スケールの散布図を作成する関数
    fastとslowの値が0の場合は1に置換してプロットする

    引数:
        df_date (DataFrame): repository, fast, slowの列を持つデータフレーム
        repo_date (list): 赤い星で表示するリポジトリ名のリスト（デフォルトは空リスト）

    戻り値:
        matplotlib.figure.Figure: 作成した図のオブジェクト
    """
    # グラフの作成
    fig = plt.figure(figsize=(7, 7))

    # まず、Other Repositoriesをプロット
    plt.scatter(
        df_date.loc[~df_date['repository'].isin(repo_date), 'fast'].replace(0, 1),
        df_date.loc[~df_date['repository'].isin(repo_date), 'slow'].replace(0, 1),
        color='black',
        marker='o',
        alpha=0.7,
        s=70,
        zorder=2,
        label='Other Repositories'
    )

    # 次に、TOP10 Repositoriesをプロット
    plt.scatter(
        df_date.loc[df_date['repository'].isin(repo_date), 'fast'].replace(0, 1),
        df_date.loc[df_date['repository'].isin(repo_date), 'slow'].replace(0, 1),
        color='blue',
        marker='o',
        s=80,
        zorder=3,
        label='TOP10 Repositories'
    )

    # データの最小値と最大値を取得（y=x線のために使用）
    # 0が1に置換されることを考慮して計算
    df_temp = df_date.copy()
    df_temp['fast'] = df_temp['fast'].replace(0, 1)
    df_temp['slow'] = df_temp['slow'].replace(0, 1)

    min_val = min(df_temp['fast'].min(), df_temp['slow'].min())
    max_val = max(df_temp['fast'].max(), df_temp['slow'].max())

    # 少し余裕を持たせる
    min_val = min_val * 0.9
    max_val = max_val * 1.1

    # y=xの直線をグレーの実線で追加
    plt.plot([min_val, max_val], [min_val, max_val], color='gray', linestyle='--', linewidth=2, label='y = x', zorder=1)

    # 軸を対数スケールに設定
    plt.xscale('log')
    plt.yscale('log')

    # 軸の目盛りのフォントサイズを大きくする
    plt.tick_params(axis='both', which='major', labelsize=20)

    # 凡例を表示
    plt.legend(fontsize=16)

    # グラフのレイアウトを調整
    plt.tight_layout()

    return fig
```

```python
import pandas as pd

def combine_keywords_dataframes(keywords_df_date_1, keywords_df_date_2, keywords_df_date_3):
    """
    3つのデータフレームからキーワードを抽出し、少なくとも1つのデータフレームに
    含まれるキーワードを抽出する関数

    Parameters:
    keywords_df_date_1, keywords_df_date_2, keywords_df_date_3:
        'keywords'カラムを持つpandas DataFrames

    Returns:
    pandas DataFrame: フィルタリングされたキーワードを含む単一のデータフレーム
    """
    # 各データフレームからキーワードを取得
    keywords_1 = set(keywords_df_date_1['keywords'])
    keywords_2 = set(keywords_df_date_2['keywords'])
    keywords_3 = set(keywords_df_date_3['keywords'])

    # すべてのキーワードを結合
    all_keywords = keywords_1.union(keywords_2, keywords_3)

    # データフレームを作成
    combined_df = pd.DataFrame(list(all_keywords), columns=['keywords'])

    # キーワードでソート
    combined_df = combined_df.sort_values('keywords').reset_index(drop=True)

    # 各日付のデータフレームに含まれているかどうかを示す列を追加
    combined_df['in_date_1'] = combined_df['keywords'].isin(keywords_1)
    combined_df['in_date_2'] = combined_df['keywords'].isin(keywords_2)
    combined_df['in_date_3'] = combined_df['keywords'].isin(keywords_3)

    # どれか1つ以上のデータフレームに含まれるキーワードをフィルタリング
    filtered_df = combined_df[
        combined_df['in_date_1'] |
        combined_df['in_date_2'] |
        combined_df['in_date_3']
    ].reset_index(drop=True)

    return filtered_df

filtered_keywords_df = combine_keywords_dataframes(keywords_df_date_1, keywords_df_date_2, keywords_df_date_3)
filtered_keywords_df.to_excel('filtered_keywords.xlsx', index=False)
```

```python
import pandas as pd

def find_unique_filtered_keywords(filtered_keywords_df, keywords_df_date_0):
    """
    filtered_keywords_dfにのみ存在するキーワードを抽出する関数

    Parameters:
    filtered_keywords_df: 'keywords'カラムを持つpandas DataFrame
    keywords_df_date_0: 'keywords'カラムを持つpandas DataFrame

    Returns:
    pandas DataFrame: filtered_keywords_dfにのみ存在するキーワードを含むデータフレーム
    """
    # 各データフレームのキーワードをセットとして取得
    filtered_keywords = set(filtered_keywords_df['keywords'])
    date_0_keywords = set(keywords_df_date_0['keywords'])

    # filtered_keywords_dfにのみ存在するキーワードを抽出
    unique_to_filtered = filtered_keywords - date_0_keywords

    # 結果のデータフレームを作成
    result_df = pd.DataFrame({
        'keywords': sorted(list(unique_to_filtered))
    })

    return result_df

# 使用例:
unique_filtered_df = find_unique_filtered_keywords(filtered_keywords_df, keywords_df_date_0)
# unique_filtered_df.to_excel('unique_filtered_keywords.xlsx', index=False)

unique_filtered_df
```

```python
import json
from collections import Counter

# JSONファイルの読み込み
with open('./drive/MyDrive/Colab Notebooks/keywords.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 全てのkeywordsを収集
all_keywords = []
for item in data:
    all_keywords.extend(item.get('keywords', []))

# keywordsの登場回数をカウント
keyword_counts = Counter(all_keywords)
# 登場回数で降順にソート
sorted_keyword_counts = sorted(keyword_counts.items(), key=lambda x: x[1], reverse=True)

# 結果の出力
for keyword, count in sorted_keyword_counts:
    print(f"{keyword}: {count}")
```
